```python
#Programa Evaluador de Creditos
#Objetivo:Demostrar la estructura de un arbol de decision anidado
 
#Score Crediticio
for score_cliente in range(400,800,50):
    deuda_mensual = 400
    ingreso_mensual= 2000
    #Proporcion de ingresos comprometidos
    relacion_deuda=deuda_mensual / ingreso_mensual
    print("Procesando cliente - Score", score_cliente ,"Relacion deuda",relacion_deuda)
    #Nodo Raiz Clientes con historial excelente
    if score_cliente >= 700:
        #Rama A
        if relacion_deuda <= 0.35:
            print("Aprobado Automaticamente")
        else:
            print("Buen score sin embargo se encuentra muy endeudado")
    #Nodo Secundario Clientes con historial moderado
    elif score_cliente >=500:
        #Rama B
        if relacion_deuda <= 0.20:
            print("Revision manual por parte del personal","\n","Score medio, bajo riesgo de deuda")
        else:
            print("Score medio y alta deuda")
    else:
        print("Rechazado Automaticamente, Riesgo financiero critico")
```

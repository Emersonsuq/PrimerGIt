```python
import re
import time
print("Bienvenido al sistema de registro itq")
print("El correo dede er minuscula y terminar en @itq.edu.ec")

# FASE 1: Validación de Correo
while True:

    correo = input("Ingrese su correo institucional: ")

    patron_correo= r"^[a-z.]+@itq\.edu\.ec$"

    if re.match(patron_correo, correo):
        print("Correo válido.")
        break

    else:
        print("Error: correo inválido.")


# FASE 2: Validación de Edad
while True:

    try:

        edad = int(input("Ingrese su edad: "))

        if edad >= 17:
            print("Edad valida.")
            break

        else:
            print("Debe tener al menos 17 años.")

    except ValueError:

        print("Error: debe ingresar un número.")



# FASE 3: Simulación Sistema
carga = 0

while carga <= 100:

    print(f"\rGenerando perfil... {carga}%", end="")

    time.sleep(0.3)

    carga += 20

print()

print(f"Bienvenido al sistema, {correo}")
```

```python
# Sistema de cobro automatizado para parqueadero.

def calcular_tarifa_parqueo(tipo_vehiculo, horas_estadia):

    if tipo_vehiculo == "moto":

        total = horas_estadia * 1.00

    else:

        total = horas_estadia * 2.00

    return total


tipo_vehiculo = input("Ingrese el tipo de vehiculo (carro o moto): ")

horas_estadia = int(input("Ingrese la cantidad de horas: "))

valor_pagar = calcular_tarifa_parqueo(tipo_vehiculo, horas_estadia)

print("El total a cobrar es: $", valor_pagar)
```

```python
import random


def simulador_vehiculo():

    try:
        distancia = int(input("Ingrese la distancia al obstáculo : "))
    except ValueError:
        print("¡Error! Ingrese un número válido.")
        return

    velocidad = random.randint(10, 120)

    if distancia <= 20 and velocidad > 60:
        accion = "FRENADO DE EMERGENCIA"

    elif distancia >= 21 and distancia <= 50:
        accion = "DESACELERACIÓN PREVENTIVA"

    else:
        accion = "MANTENER CURSO NORMAL"

    print("\n------ REPORTE ------")
    print("========================")
    print(f"Distancia: {distancia} metros")
    print(f"Velocidad: {velocidad} km/h")
    print(f"Acción: {accion}")
    print("=======================")


def menu_principal():

    while True:

        print("\n===== SISTE DE CONTROL AUTOMOTRIZ =====")
        print("1. Ejecutar simulación")
        print("2. Salir")

        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            simulador_vehiculo()

        elif opcion == "2":
            print("Programa finalizado.")
            break

        else:
            print("Opción inválida.")

menu_principal()
```

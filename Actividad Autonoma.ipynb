```python
# Autores: EMERSON GONZALO SUQUILLO CUZCO
#DATOS DEL SISTEMA (SUCIOS Y REPETIDOS)
acceso_logica = ["JuanP", "AnaM", "LuisG", "AnaM", "CarlosT", "JuanP", "SofiaR", "AnaM"]
acceso_web = ["CarlosT", "SofiaR", "ElenaV", "MarioL", "SofiaR", "JuanP", "MarioL"]

#RETO 1: PURGA DE DATOS (Deduplicación)
set_logica = set(acceso_logica)
set_web = set(acceso_web)


#RETO 2: INTERSECCIÓN 
ambas_materias = set_logica & set_web

#RETO 3: DIFERENCIA
solo_logica = set_logica - set_web

# RETO 4: UNIÓN 

lista_maestra = set_logica | set_web


# REPORTE FINAL

print("------ REPORTE ------")

print(f"Total de estudiantes únicos en Lógica: {len(set_logica)}")
print(f"Estudiantes de Lógica: {set_logica}\n\n")
print(f"Total de estudiantes únicos en Desarrollo Web: {len(set_web)}")
print(f"Estudiantes de Desarrollo Web: {set_web}\n")
print(f"Estudiantes que cursan ambas materias: {ambas_materias}")
print(f"Estudiantes solo de Lógica: {solo_logica}")
print(f"Lista maestra de estudiantes: {lista_maestra}")
print(f"Total de estudiantes registrados: {len(lista_maestra)}")
```
